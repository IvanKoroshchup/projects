```python
#–ü–æ—Å—á–∏—Ç–∞–µ–º —Å —Ä–∞–∑–±–∏–≤–∫–æ–π –ø–æ –º–µ—Å—è—Ü–∞–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–¥–µ–ª–∞–Ω–Ω—ã—Ö –∑–≤–æ–Ω–∫–æ–≤ –∫–∞–∂–¥—ã–º –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–º
data_calls_new['month'] = data_calls_new['call_date'].dt.month
data_users_calls = data_users.merge(data_calls_new, how='outer', on='user_id')
print('===========================================================================================================')
print('–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–¥–µ–ª–∞–Ω–Ω—ã—Ö –∑–≤–æ–Ω–∫–æ–≤')
count_calls = data_users_calls.pivot_table(index = ['user_id','tariff'],columns = 'month',values = 'duration',aggfunc ='count',fill_value = 0)
display(count_calls)

#–ü–æ—Å—á–∏—Ç–∞–µ–º —Å —Ä–∞–∑–±–∏–≤–∫–æ–π –ø–æ –º–µ—Å—è—Ü–∞–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏–∑—Ä–∞—Å—Ö–æ–¥–æ–≤–∞–Ω–Ω—ã—Ö –º–∏–Ω—É—Ç –∫–∞–∂–¥—ã–º –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–º
print('===========================================================================================================')
print('–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏–∑—Ä–∞—Å—Ö–æ–¥–æ–≤–∞–Ω–Ω—ã—Ö –º–∏–Ω—É—Ç')
sum_calls = data_users_calls.pivot_table(index = ['user_id','tariff'],columns = 'month',values = 'duration',aggfunc ='sum',fill_value = 0)
display(sum_calls)

#–ü–æ—Å—á–∏—Ç–∞–µ–º —Å —Ä–∞–∑–±–∏–≤–∫–æ–π –ø–æ –º–µ—Å—è—Ü–∞–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ—Ç–ø—Ä–∞–≤–ª–µ–Ω–Ω—ã—Ö –°–ú–°
data_messages['month'] = data_messages['message_date'].dt.month
data_users_messages = data_users.merge(data_messages, how='outer', on='user_id')
print('===========================================================================================================')
print('–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ—Ç–ø—Ä–∞–≤–ª–µ–Ω–Ω—ã—Ö —Å–æ–æ–±—â–µ–Ω–∏–π')
count_messages = data_users_messages.pivot_table(index = ['user_id','tariff'],columns = 'month',values = 'id',aggfunc ='count',fill_value = 0)
display(count_messages)

#–ü–æ—Å—á–∏—Ç–∞–µ–º –æ–±—ä–µ–º –∏–∑—Ä–∞—Å—Ö–æ–¥–æ–≤–∞–Ω–Ω–æ–≥–æ –∏–Ω—Ç–µ—Ä–Ω–µ—Ç —Ç—Ä–∞—Ñ—Ñ–∏–∫–∞ —Å —Ä–∞–∑–±–∏–≤–∫–æ–π –ø–æ –º–µ—Å—è—Ü–∞–º
data_internet['month'] = data_internet['session_date'].dt.month
data_users_internet = data_users.merge(data_internet, how='outer', on='user_id')
print('===========================================================================================================')
print('–æ–±—ä–µ–º –∏–∑—Ä–∞—Å—Ö–æ–¥–æ–≤–∞–Ω–Ω—ã—Ö –º–µ–≥–∞–±–∞–π—Ç–æ–≤')
sum_internet = data_users_internet.pivot_table(index = ['user_id','tariff'],columns = 'month',values = 'mb_used',aggfunc ='sum',fill_value = 0)
display(sum_internet)

#–î–æ–±–∞–≤–∏–º –¥–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –∫–æ–ª–æ–Ω–∫–∏ –º–µ—Å—è—Ü–∞ –ø–æ–¥–∫–ª—é—á–µ–Ω–∏—è (–∏ –æ—Ç–∫–ª—é—á–µ–Ω–∏—è) —Ç–∞—Ä–∏—Ñ–∞
data_users['reg_month'] = data_users['reg_date'].dt.month
data_users['churn_month'] = data_users['churn_date'].dt.month
data_users_reg = data_users[['user_id','churn_month','reg_month']]
```

  ### –ü–æ—Å—á–∏—Ç–∞–µ–º –¥–æ—Ö–æ–¥ —Å –∫–∞–∂–¥–æ–≥–æ –∫–ª–∏–µ–Ω—Ç–∞ –ø–æ –º–µ—Å—è—Ü–∞–º

```python
#–ü–æ—Å—á–∏—Ç–∞–µ–º –¥–æ—Ö–æ–¥ —Å –∫–∞–∂–¥–æ–≥–æ –∫–ª–∏–µ–Ω—Ç–∞ –ø–æ –º–µ—Å—è—Ü–∞–º
#–°–æ–±–µ—Ä–µ–º –æ–¥–Ω—É —Ç–∞–±–ª–∏—Ü—É –æ–±—ä–µ–¥–µ–Ω–∏–≤ —Ç–∞–±–ª–∏—Ü—ã (—Å–æ–æ–±—â–µ–Ω–∏—è + –∑–≤–æ–Ω–∫–∏ + –º–µ–≥–∞–±–∞–π—Ç—ã + –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏(–∫–æ–ª–æ–Ω–∫–∏ –¥–∞—Ç–∞ –ø–æ–¥–∫–ª—é—á–µ–Ω–∏—è/–æ—Ç–∫–ª—é—á–µ–Ω–∏—è))
profit_user = sum_calls.join(count_messages, how = 'outer',lsuffix = '_calls', rsuffix='_sms')
sum_internet = sum_internet.add_suffix('_mb')
profit_user = profit_user.join(sum_internet, how = 'outer')
profit_user.reset_index(inplace=True)
profit_user = profit_user.merge(data_users_reg, on = 'user_id', how = 'left')

#–ó–∞–º–µ–Ω–∏–º –æ—Ç—Å—É—Ç—Å—Ç–≤—É—é—â–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ –∫–æ–ª–æ–Ω–∫–µ –æ–∫–æ–Ω—á–∞–Ω–∏—è —Ç–∞—Ä–∏—Ñ–Ω–æ–≥–æ –ø–ª–∞–Ω–∞ (—ç—Ç–æ –Ω—É–∂–Ω–æ –¥–ª—è –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ—Å—Ç–∏ —Ñ—É–Ω–∫—Ü–∏–∏)
#–í –∑–Ω–∞—á–µ–Ω–∏–µ valid –¥–ª—è —Ñ—É–Ω–∫—Ü–∏–∏ –º–æ–∂–Ω–æ –ø–æ—Å—Ç–∞–≤–∏—Ç—å 1000 –Ω–∞–ø—Ä–∏–º–µ—Ä, —Ç.–∫. –Ω–∞—à–∏ –¥–∞–Ω–Ω—ã–µ –≤—Å–µ —Å–æ–±—Ä–∞–Ω—ã –≤ 2018–≥, –ø—É—Å—Ç—å –±—É–¥–µ—Ç —Ä–∞–≤–Ω—ã–º 12
valid = 12
profit_user = profit_user.fillna({'churn_month':valid})
profit_user = profit_user.fillna(0)

#–î–ª—è –ø—Ä–æ—Å—á–µ—Ç–∞ –¥–æ—Ö–æ–¥–Ω–æ—Å—Ç–∏ –ø–æ –º–µ—Å—è—Ü–∞–º –Ω–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏–∏
def profit_jan (row):
    over_calls = 0
    over_messages = 0
    over_mb = 0
    calls = row['1.0_calls']
    sms = row['1.0_sms']
    mb = row['1.0_mb']
    month_numb = 1
    
    if row['tariff'] == "smart":
        tariff_params = data_tariffs.loc[0]
    elif row['tariff'] == "ultra":
        tariff_params = data_tariffs.loc[1]     
    
    if row['reg_month'] > month_numb or row['churn_month'] < month_numb:
        result = 0
    else:
        if calls > tariff_params['minutes_included']:
            over_calls = (calls - tariff_params['minutes_included'])*tariff_params['rub_per_minute']
        else:
            over_calls = 0
        if sms > tariff_params['messages_included']:
            over_messages = (sms - tariff_params['messages_included'])*tariff_params['rub_per_message']
        else:
            over_messages = 0
        if mb > tariff_params['mb_per_month_included']:
            over_mb = (mb - tariff_params['mb_per_month_included'])*(tariff_params['rub_per_gb']/1024)
        else:
            over_mb = 0
        result = over_calls + over_messages + over_mb + tariff_params['rub_monthly_fee']
        return result
profit_user['profit_jan'] = profit_user.apply(profit_jan, axis = 1)
```

```python
#–†–∞–∑–æ–±—ä–µ–º —Ç–∞–±–ª–∏—Ü—ã –ø–æ —Ç–∞—Ä–∏—Ñ–∞–º 
#–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –º–∏–Ω—É—Ç –∑–≤–æ–Ω–∫–æ–≤ - —Ç–∞—Ä–∏—Ñ–∞ "—Å–º–∞—Ä—Ç"
sum_calls_smart = sum_calls.query('tariff=="smart"')
mean_calls_smart = sum_calls_smart.mean()

#–ü–æ—Å—á–∏—Ç–∞–µ–º –¥–∏—Å–ø–µ—Ä—Å–∏—é –∏ —Å—Ä–µ–¥–Ω–∏–π —Ä–∞–∑–±—Ä–æ—Å
varianse_calls_smart = np.var(sum_calls_smart, ddof=1)
standard_deviation_calls_smart =  np.sqrt(varianse_calls_smart)

#–ü–æ—Å—Ç—Ä–æ–∏–º –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É —Å—Ä–µ–¥–Ω–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∑–≤–æ–Ω–∫–æ–≤ 
ax = mean_calls_smart.plot(kind = 'bar', figsize = (12,3),label = '—Ç–∞—Ä–∏—Ñ —Å–º–∞—Ä—Ç', legend = True, title = '–ó–≤–æ–Ω–∫–∏', alpha = 0.5, color = 'red')
#===========================================================================================================================

#–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –º–∏–Ω—É—Ç –∑–≤–æ–Ω–∫–æ–≤ - —Ç–∞—Ä–∏—Ñ–∞ "—É–ª—å—Ç—Ä–∞"
sum_calls_ultra = sum_calls.query('tariff=="ultra"')
mean_calls_ultra = sum_calls_ultra.mean()

#–ü–æ—Å—á–∏—Ç–∞–µ–º –¥–∏—Å–ø–µ—Ä—Å–∏—é –∏ —Å—Ä–µ–¥–Ω–∏–π —Ä–∞–∑–±—Ä–æ—Å
varianse_calls_ultra = np.var(sum_calls_ultra, ddof=1)
standard_deviation_calls_ultra =  np.sqrt(varianse_calls_ultra)

#–ü–æ—Å—Ç—Ä–æ–∏–º –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∑–≤–æ–Ω–∫–æ–≤(—Å–æ–µ–¥–∏–Ω–∏–º —Å –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º–æ–π —Ç–∞—Ä–∏—Ñ–∞ "—Å–º–∞—Ä—Ç")
mean_calls_ultra.plot(ax=ax, kind = 'bar',label = '—Ç–∞—Ä–∏—Ñ —É–ª—å—Ç—Ä–∞', legend = True, grid = True, alpha = 0.5)
plt.xlabel('–º–µ—Å—è—Ü')
plt.ylabel('—Å—Ä–µ–¥–Ω–µ–µ  –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ  –º–∏–Ω—É—Ç')
plt.show()

#–ü–æ—Å—Ç—Ä–æ–∏–º –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è –ø–æ —Å—Ä–µ–¥–Ω–µ–º—É –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –º–∏–Ω—É—Ç (—Å–æ–µ–¥–∏–Ω–∏–º –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã "—Å–º–∞—Ä—Ç" –∏ "—É–ª—å—Ç—Ä–∞" –Ω–∞ –æ–¥–Ω–æ–π)
ax = standard_deviation_calls_smart.plot(kind = 'bar', figsize = (12,3),label = '—Ç–∞—Ä–∏—Ñ —Å–º–∞—Ä—Ç', legend = True, title = '–ó–≤–æ–Ω–∫–∏ - —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ —Å—Ä–µ–¥–Ω–µ–≥–æ', alpha = 1, color = 'green')
standard_deviation_calls_ultra.plot(ax=ax, kind = 'bar',label = '—Ç–∞—Ä–∏—Ñ —É–ª—å—Ç—Ä–∞', legend = True, grid = True, alpha = 0.8, color = 'yellow')
plt.xlabel('–º–µ—Å—è—Ü')
plt.ylabel('–æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –ø–æ —Å—Ä–µ–¥–Ω–µ–º—É –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –º–∏–Ω—É—Ç')
plt.show()
#===========================================================================================================================

#–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ—Ç–ø—Ä–∞–≤–ª–µ–Ω–Ω—ã—Ö —Å–æ–æ–±—â–µ–Ω–∏–π
#–¢–∞—Ä–∏—Ñ "—Å–º–∞—Ä—Ç" - —Å—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ—Ç–ø—Ä–∞–≤–ª–µ–Ω–Ω—ã—Ö —Å–æ–æ–±—â–µ–Ω–∏–π"
count_messages_smart = count_messages.query('tariff=="smart"')
mean_messages_smart = count_messages_smart.mean()

#–ü–æ—Å—á–∏—Ç–∞–µ–º –¥–∏—Å–ø–µ—Ä—Å–∏—é –∏ —Å—Ä–µ–¥–Ω–∏–π —Ä–∞–∑–±—Ä–æ—Å
varianse_messages_smart = np.var(count_messages_smart, ddof=1)
standard_deviation_messages_smart =  np.sqrt(varianse_messages_smart)

#–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ —Å—Ä–µ–¥–Ω–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –æ—Ç–ø—Ä–∞–≤–ª–µ–Ω–Ω—ã—Ö —Å–æ–æ–±—â–µ–Ω–∏–π
ax = mean_messages_smart.plot(kind = 'bar', figsize = (12,3),label = '—Ç–∞—Ä–∏—Ñ —Å–º–∞—Ä—Ç', legend = True, title = '–°–æ–æ–±—â–µ–Ω–∏—è', alpha = 0.5, color = 'red')

#–¢–∞—Ä–∏—Ñ "—É–ª—å—Ç—Ä–∞" - —Å—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ—Ç–ø—Ä–∞–≤–ª–µ–Ω–Ω—ã—Ö —Å–æ–æ–±—â–µ–Ω–∏–π"
count_messages_ultra = count_messages.query('tariff=="ultra"')
mean_messages_ultra = count_messages_ultra.mean()

#–ü–æ—Å—á–∏—Ç–∞–µ–º –¥–∏—Å–ø–µ—Ä—Å–∏—é –∏ —Å—Ä–µ–¥–Ω–∏–π —Ä–∞–∑–±—Ä–æ—Å
varianse_messages_ultra = np.var(count_messages_ultra, ddof=1)
standard_deviation_messages_ultra =  np.sqrt(varianse_messages_ultra)

#–ü–æ—Å—Ç—Ä–æ–∏–º –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∑–≤–æ–Ω–∫–æ–≤(—Å–æ–µ–¥–∏–Ω–∏–º —Å –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º–æ–π —Ç–∞—Ä–∏—Ñ–∞ "—Å–º–∞—Ä—Ç")
mean_messages_ultra.plot(ax=ax, kind = 'bar',label = '—Ç–∞—Ä–∏—Ñ —É–ª—å—Ç—Ä–∞', legend = True, grid = True, alpha = 0.5)
plt.xlabel('–º–µ—Å—è—Ü')
plt.ylabel('—Å—Ä–µ–¥–Ω–µ–µ  –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ  —Å–æ–æ–±—â–µ–Ω–∏–π')
plt.show()

#–ü–æ—Å—Ç—Ä–æ–∏–º –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è –ø–æ —Å—Ä–µ–¥–Ω–µ–º—É –∫–æ–ª–∏—á–µ—Å—Ç–≤—É —Å–æ–æ–±—â–µ–Ω–∏–π (—Å–æ–µ–¥–∏–Ω–∏–º –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã "—Å–º–∞—Ä—Ç" –∏ "—É–ª—å—Ç—Ä–∞" –Ω–∞ –æ–¥–Ω–æ–π)
ax = standard_deviation_messages_smart.plot(kind = 'bar', figsize = (12,3),label = '—Ç–∞—Ä–∏—Ñ —Å–º–∞—Ä—Ç', legend = True, title = '–°–æ–æ–±—â–µ–Ω–∏—è - —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ —Å—Ä–µ–¥–Ω–µ–≥–æ', alpha = 0.8, color = 'green')
standard_deviation_messages_ultra.plot(ax=ax, kind = 'bar',label = '—Ç–∞—Ä–∏—Ñ —É–ª—å—Ç—Ä–∞', legend = True, grid = True, alpha = 0.8, color = 'yellow')
plt.xlabel('–º–µ—Å—è—Ü')
plt.ylabel('–æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –ø–æ —Å—Ä–µ–¥–Ω–µ–º—É –∫–æ–ª–∏—á–µ—Å—Ç–≤—É —Å–æ–æ–±—â–µ–Ω–∏–π')
plt.show()
#===========================================================================================================================

#–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏–∑—Ä–∞—Å—Ö–æ–¥–æ–≤–∞–Ω–Ω—ã—Ö –º–µ–≥–∞–±–∞–π—Ç
#–¢–∞—Ä–∏—Ñ "—Å–º–∞—Ä—Ç" - —Å—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏–∑—Ä–∞—Å—Ö–æ–¥–æ–≤–∞–Ω–Ω—ã—Ö –º–µ–≥–∞–±–∞–π—Ç–æ–≤"
sum_internet_smart = sum_internet.query('tariff=="smart"')
mean_internet_smart = sum_internet_smart.mean()

#–ü–æ—Å—á–∏—Ç–∞–µ–º –¥–∏—Å–ø–µ—Ä—Å–∏—é –∏ —Å—Ä–µ–¥–Ω–∏–π —Ä–∞–∑–±—Ä–æ—Å
varianse_internet_smart = np.var(sum_internet_smart, ddof=1)
standard_deviation_internet_smart =  np.sqrt(varianse_internet_smart)

#–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ —Å—Ä–µ–¥–Ω–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∏–∑—Ä–∞—Å—Ö–æ–¥–æ–≤–∞–Ω–Ω—ã—Ö –º–µ–≥–∞–±–∞–π—Ç–æ–≤
ax = mean_internet_smart.plot(kind = 'bar', figsize = (12,3),label = '—Ç–∞—Ä–∏—Ñ —Å–º–∞—Ä—Ç', legend = True, title = '–ú–µ–≥–∞–±–∞–π—Ç—ã', alpha = 0.5, color = 'red')

#–¢–∞—Ä–∏—Ñ "—É–ª—å—Ç—Ä–∞" - —Å—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏–∑—Ä–∞—Å—Ö–æ–¥–æ–≤–∞–Ω–Ω—ã—Ö –º–µ–≥–∞–±–∞–π—Ç–æ–≤"
sum_internet_ultra = sum_internet.query('tariff=="ultra"')
mean_internet_ultra = sum_internet_ultra.mean()

#–ü–æ—Å—á–∏—Ç–∞–µ–º –¥–∏—Å–ø–µ—Ä—Å–∏—é –∏ —Å—Ä–µ–¥–Ω–∏–π —Ä–∞–∑–±—Ä–æ—Å
varianse_internet_ultra = np.var(sum_internet_ultra, ddof=1)
standard_deviation_internet_ultra =  np.sqrt(varianse_internet_ultra)

#–ü–æ—Å—Ç—Ä–æ–∏–º –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É –∏–∑—Ä–∞—Å—Ö–æ–¥–æ–≤–∞–Ω–Ω—ã—Ö –º–µ–≥–∞–±–∞–π—Ç–æ–≤(—Å–æ–µ–¥–∏–Ω–∏–º —Å –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º–æ–π —Ç–∞—Ä–∏—Ñ–∞ "—Å–º–∞—Ä—Ç")
mean_internet_ultra.plot(ax=ax, kind = 'bar',label = '—Ç–∞—Ä–∏—Ñ —É–ª—å—Ç—Ä–∞', legend = True, grid = True, alpha = 0.5)
plt.xlabel('–º–µ—Å—è—Ü')
plt.ylabel('—Å—Ä–µ–¥–Ω–µ–µ  –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ  –ø–æ—Ç—Ä–∞—á–µ–Ω–Ω—ã—Ö  –º–µ–≥–∞–±–∞–π—Ç')
plt.show()

#–ü–æ—Å—Ç—Ä–æ–∏–º –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è –ø–æ —Å—Ä–µ–¥–Ω–µ–º—É –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –∏–∑—Ä–∞—Å—Ö–æ–¥–æ–≤–∞–Ω–Ω—ã—Ö –º–µ–≥–∞–±–∞–π—Ç–æ–≤(—Å–æ–µ–¥–∏–Ω–∏–º –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã "—Å–º–∞—Ä—Ç" –∏ "—É–ª—å—Ç—Ä–∞" –Ω–∞ –æ–¥–Ω–æ–π)
ax = standard_deviation_internet_smart.plot(kind = 'bar', figsize = (12,3),label = '—Ç–∞—Ä–∏—Ñ —Å–º–∞—Ä—Ç', legend = True, title = '–ò–Ω—Ç–µ—Ä–Ω–µ—Ç —Ç—Ä–∞—Ñ—Ñ–∏–∫ - —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ —Å—Ä–µ–¥–Ω–µ–≥–æ', alpha = 0.8, color = 'green')
standard_deviation_internet_ultra.plot(ax=ax, kind = 'bar',label = '—Ç–∞—Ä–∏—Ñ —É–ª—å—Ç—Ä–∞', legend = True, grid = True, alpha = 0.8, color = 'yellow')
plt.xlabel('–º–µ—Å—è—Ü')
plt.ylabel('–æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –ø–æ —Å—Ä–µ–¥–Ω–µ–º—É –æ–±—ä–µ–º—É —Ç—Ä–∞—Ñ—Ñ–∏–∫–∞')
plt.show()
```

–ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –±—É–¥–µ—Ç –∑–≤—É—á–∞—Ç—å —Ç–∞–∫: –°—Ä–µ–¥–Ω–µ–µ –¥–≤—É—Ö –≤—ã–±–æ—Ä–æ–∫ —Ä–∞–∑–ª–∏—á–∞–µ—Ç—Å—è, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ –≤—ã—Ä—É—á–∫–∞ –æ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≥–æ—Ä–æ–¥—ã –ú–æ—Å–∫–≤—ã –∏ –¥—Ä—É–≥–∏—Ö –≥–æ—Ä–æ–¥–æ–≤ –¥–ª—è –∫–æ–º–ø–∞–Ω–∏–∏ —Ä–∞–∑–ª–∏—á–∞–µ—Ç—Å—è

```python
#–†–∞–∑–æ–±—å–µ–º –Ω–∞ –≥–æ—Ä–æ–¥–∞ –ú–æ—Å–∫–≤–∞/–¥—Ä—É–≥–∏–µ –≥–æ—Ä–æ–¥–∞
    #–°–æ–µ–¥–∏–Ω–∏–º —Ç–∞–±–ª–∏—Ü—ã –≥–æ–¥–æ–≤–æ–π –≤—ã—Ä—É—á–∫–∏ –∏ —Ç–∞–±–ª–∏—Ü—É –¥–∞–Ω–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π(–Ω—É–∂–Ω–∞ –∫–æ–ª–æ–Ω–∫–∞ —Å –≥–æ—Ä–æ–¥–æ–º)
data_users_city = data_users[['user_id','city']]
profit_user_year_city = profit_user_year.merge(data_users_city, on = 'user_id', how = 'outer')

#–°–¥–µ–ª–∞–µ–º –≤—ã–±–æ—Ä–∫—É —Å –≥–æ–¥–æ–≤–æ–π –≤—ã—Ä—É—á–∫–æ–π –ø–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è–º –∏–∑ –ú–æ—Å–∫–≤—ã
profit_user_year_city_moscow = profit_user_year_city\
                .query('city == "–ú–æ—Å–∫–≤–∞"')\
                .drop(['user_id','tariff','city'], axis = 1)\
                .sum(axis = 'columns')

#–°–¥–µ–ª–∞–µ–º –≤—ã–±–æ—Ä–∫—É —Å –≥–æ–¥–æ–≤–æ–π –≤—ã—Ä—É—á–∫–æ–π –ø–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è–º –∏–∑ –¥—Ä—É–≥–∏—Ö –≥–æ—Ä–æ–¥–æ–≤
profit_user_year_city_no_moscow = profit_user_year_city\
                .query('city != "–ú–æ—Å–∫–≤–∞"')\
                .drop(['user_id','tariff','city'], axis = 1)\
                .sum(axis = 'columns')

#–°—Ä–∞–≤–Ω–∏–º —Å—Ä–µ–¥–Ω–µ–µ –≤—ã–±–æ—Ä–æ–∫ –≤—ã—Ä—É—á–∫–∞ –ú–æ—Å–∫–≤–∞/–¥—Ä—É–≥–∏–µ –≥–æ—Ä–æ–¥–∞
    #–ü—Ä–æ–≤–µ—Ä–∏–º –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç—å —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –≤—ã–±–æ—Ä–æ–∫ –ú–æ—Å–∫–≤–∞/ –¥—Ä—É–≥–∏–µ –≥–æ—Ä–æ–¥–∞ —Ç–µ—Å—Ç–æ–º –®–∞–ø–∏—Ä–æ-–£–∏–ª–∫–∞
stat_moscov,pvalue_mocsow = st.shapiro(profit_user_year_city_moscow)
stat_no_moscov,pvalue_no_mocsow = st.shapiro(profit_user_year_city_no_moscow)
alpha = .05

if pvalue_mocsow > alpha:
    print('–¥–∞–Ω–Ω—ã–µ –≤ –≤—ã–±–æ—Ä–∫–µ –≥. –ú–æ—Å–∫–≤–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –Ω–æ—Ä–º–∞–ª—å–Ω–æ (—Ç–µ—Å—Ç –®–∞–ø–∏—Ä–æ-–£–∏–ª–∫–∞)')
else:
    print('–¥–∞–Ω–Ω—ã–µ –≤ –≤—ã–±–æ—Ä–∫–µ –≥. –ú–æ—Å–∫–≤–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –Ω–µ –Ω–æ—Ä–º–∞–ª—å–Ω–æ (—Ç–µ—Å—Ç –®–∞–ø–∏—Ä–æ-–£–∏–ª–∫–∞)')

if pvalue_no_mocsow > alpha:
    print('–¥–∞–Ω–Ω—ã–µ –≤ –≤—ã–±–æ—Ä–∫–µ –ø–æ –≥–æ—Ä–æ–¥–∞–º –∫—Ä–æ–º–µ –ú–æ—Å–∫–≤—ã —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –Ω–æ—Ä–º–∞–ª—å–Ω–æ (—Ç–µ—Å—Ç –®–∞–ø–∏—Ä–æ-–£–∏–ª–∫–∞)')
else:
    print('–¥–∞–Ω–Ω—ã–µ –≤ –≤—ã–±–æ—Ä–∫–µ –ø–æ –≥–æ—Ä–æ–¥–∞–º –∫—Ä–æ–º–µ –ú–æ—Å–∫–≤—ã —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –Ω–µ –Ω–æ—Ä–º–∞–ª—å–Ω–æ (—Ç–µ—Å—Ç –®–∞–ø–∏—Ä–æ-–£–∏–ª–∫–∞)')    
print()

#–í–∏–¥–∏–º —á—Ç–æ –æ–±–µ –≤—ã–±–æ—Ä–∫–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –Ω–µ –Ω–æ—Ä–º–∞–ª—å–Ω–æ, –∏—Å–ø–æ–ª—å–∑—É–µ–º —Ç–µ—Å—Ç –õ–µ–≤–µ–Ω–∞ –¥–ª—è –æ—Ü–µ–Ω–∫–∏ —Ä–∞–≤–µ–Ω—Å—Ç–≤–∞ –¥–∏—Å–ø–µ—Ä—Å–∏–π
alpha = .05
stat, pval = st.levene(profit_user_year_city_moscow, profit_user_year_city_no_moscow)
if pval > alpha:
    print('–¥–∏—Å–ø–µ—Ä—Å–∏—è –≤—ã–±–æ—Ä–æ–∫ –ú–æ—Å–∫–≤–∞ / –¥—Ä—É–≥–∏–µ –≥–æ—Ä–æ–¥–∞ - —Ä–∞–≤–Ω–∞ (—Ç–µ—Å—Ç –õ–µ–≤–µ–Ω–∞)')
else:
    print('–¥–∏—Å–ø–µ—Ä—Å–∏—è –≤—ã–±–æ—Ä–æ–∫ –ú–æ—Å–∫–≤–∞ / –¥—Ä—É–≥–∏–µ –≥–æ—Ä–æ–¥–∞ - –Ω–µ —Ä–∞–≤–Ω–∞ (—Ç–µ—Å—Ç –õ–µ–≤–µ–Ω–∞)')    
print()

#–ó–Ω–∞—è, —á—Ç–æ –¥–∏—Å–ø–µ—Ä—Å–∏–∏ —Ä–∞–≤–Ω—ã, –≤–ø–∏—Å—ã–≤–∞–µ–º –ø–∞—Ä–∞–º–µ—Ç—Ä equal_var
alpha = .05
results = st.ttest_ind(profit_user_year_city_moscow, profit_user_year_city_no_moscow, equal_var = True)

if (results.pvalue < alpha):
    print("–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É (–≤—ã—Ä—É—á–∫–∞ –æ—Ç –∞–±–æ–Ω–µ–Ω—Ç–æ–≤ –ú–æ—Å–∫–≤—ã –∏ –¥—Ä—É–≥–∏—Ö –≥–æ—Ä–æ–¥–∞—Ö –æ–¥–∏–Ω–∞–∫–æ–≤–∞—è)")
else:
    print("–ù–µ –ø–æ–ª—É—á–∏–ª–æ—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É (–≤—ã—Ä—É—á–∫–∞ –æ—Ç –∞–±–æ–Ω–µ–Ω—Ç–æ–≤ –ú–æ—Å–∫–≤—ã –∏ –¥—Ä—É–≥–∏—Ö –≥–æ—Ä–æ–¥–æ–≤ –æ–¥–∏–Ω–∞–∫–æ–≤–∞—è)") 
print('============================================================================================')


#–°—Ä–∞–≤–Ω–∏–º —Å—Ä–µ–¥–Ω–µ–µ –≤—ã–±–æ—Ä–æ–∫ –≤—ã—Ä—É—á–∫–∞ —Ç–∞—Ä–∏—Ñ–æ–≤ —Å–º–∞—Ä—Ç/—É–ª—å—Ç—Ä–∞
    #–£–∑–Ω–∞–µ–º –≥–æ–¥–æ–≤—É—é –≤—ã—Ä—É—á–∫—É –ø–æ –∫–∞–∂–¥–æ–º—É –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—é —Ç–∞—Ä–∏—Ñ–æ–≤(–≤—ã–±–æ—Ä–∫–∞ –ø–æ —Ç–∞—Ä–∏—Ñ—É "—Å–º–∞—Ä—Ç" –∏ —É–ª—å—Ç—Ä–∞ —É –Ω–∞—Å —É–∂–µ –µ—Å—Ç—å)
profit_user_year_smart_sum = profit_user_year_smart.sum(axis = 'columns')
profit_user_year_ultra_sum = profit_user_year_ultra.sum(axis = 'columns') 
    
    #–ü—Ä–æ–≤–µ—Ä–∏–º –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç—å —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –≤—ã–±–æ—Ä–æ–∫ –ú–æ—Å–∫–≤–∞/ –¥—Ä—É–≥–∏–µ –≥–æ—Ä–æ–¥–∞ —Ç–µ—Å—Ç–æ–º –®–∞–ø–∏—Ä–æ-–£–∏–ª–∫–∞
stat_smart,pvalue_smart = st.shapiro(profit_user_year_smart_sum)
stat_ultra,pvalue_ultra = st.shapiro(profit_user_year_ultra_sum)
alpha = .05
    
if pvalue_smart > alpha:
    print('–¥–∞–Ω–Ω—ã–µ –≤ –≤—ã–±–æ—Ä–∫–µ —Ç–∞—Ä–∏—Ñ "—Å–º–∞—Ä—Ç" —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –Ω–æ—Ä–º–∞–ª—å–Ω–æ (—Ç–µ—Å—Ç –®–∞–ø–∏—Ä–æ-–£–∏–ª–∫–∞)')
else:
    print('–¥–∞–Ω–Ω—ã–µ –≤ –≤—ã–±–æ—Ä–∫–µ —Ç–∞—Ä–∏—Ñ "—Å–º–∞—Ä—Ç" —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –Ω–µ –Ω–æ—Ä–º–∞–ª—å–Ω–æ (—Ç–µ—Å—Ç –®–∞–ø–∏—Ä–æ-–£–∏–ª–∫–∞)')

if pvalue_ultra > alpha:
    print('–¥–∞–Ω–Ω—ã–µ –≤ –≤—ã–±–æ—Ä–∫–µ —Ç–∞—Ä–∏—Ñ "—É–ª—å—Ç—Ä–∞" —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –Ω–æ—Ä–º–∞–ª—å–Ω–æ (—Ç–µ—Å—Ç –®–∞–ø–∏—Ä–æ-–£–∏–ª–∫–∞)')
else:
    print('–¥–∞–Ω–Ω—ã–µ –≤ –≤—ã–±–æ—Ä–∫–µ —Ç–∞—Ä–∏—Ñ "—É–ª—å—Ç—Ä–∞" —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –Ω–µ –Ω–æ—Ä–º–∞–ª—å–Ω–æ (—Ç–µ—Å—Ç –®–∞–ø–∏—Ä–æ-–£–∏–ª–∫–∞)')    
print()

#–í–∏–¥–∏–º —á—Ç–æ –æ–±–µ –≤—ã–±–æ—Ä–∫–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –Ω–µ –Ω–æ—Ä–º–∞–ª—å–Ω–æ, –∏—Å–ø–æ–ª—å–∑—É–µ–º —Ç–µ—Å—Ç –õ–µ–≤–µ–Ω–∞ –¥–ª—è –æ—Ü–µ–Ω–∫–∏ —Ä–∞–≤–µ–Ω—Å—Ç–≤–∞ –¥–∏—Å–ø–µ—Ä—Å–∏–π
alpha = .05
stat_tariffs, pval_tariffs = st.levene(profit_user_year_smart_sum, profit_user_year_ultra_sum)
if pval_tariffs > alpha:
    print('–¥–∏—Å–ø–µ—Ä—Å–∏—è –≤—ã–±–æ—Ä–æ–∫ –ú–æ—Å–∫–≤–∞ / –¥—Ä—É–≥–∏–µ –≥–æ—Ä–æ–¥–∞ - —Ä–∞–≤–Ω–∞ (—Ç–µ—Å—Ç –õ–µ–≤–µ–Ω–∞)')
else:
    print('–¥–∏—Å–ø–µ—Ä—Å–∏—è –≤—ã–±–æ—Ä–æ–∫ –ú–æ—Å–∫–≤–∞ / –¥—Ä—É–≥–∏–µ –≥–æ—Ä–æ–¥–∞ - –Ω–µ —Ä–∞–≤–Ω–∞ (—Ç–µ—Å—Ç –õ–µ–≤–µ–Ω–∞)')    
print()    
    
#–ó–Ω–∞—è, —á—Ç–æ –¥–∏—Å–ø–µ—Ä—Å–∏–∏ —Ä–∞–≤–Ω—ã, –≤–ø–∏—Å—ã–≤–∞–µ–º –ø–∞—Ä–∞–º–µ—Ç—Ä equal_var
alpha = .05
results = st.ttest_ind(profit_user_year_smart_sum, profit_user_year_ultra_sum, equal_var = False)
if (results.pvalue < alpha):
    print("–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É: (–≤—ã—Ä—É—á–∫–∞ –æ—Ç –∞–±–æ–Ω–µ–Ω—Ç–æ–≤ —Ä–∞–∑–Ω—ã—Ö —Ç–∞—Ä–∏—Ñ–Ω—ã—Ö –ø–ª–∞–Ω–æ–≤ –æ–¥–∏–Ω–∞–∫–æ–≤–∞—è)")
else:
    print("–ù–µ –ø–æ–ª—É—á–∏–ª–æ—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É: (–≤—ã—Ä—É—á–∫–∞ –æ—Ç –∞–±–æ–Ω–µ–Ω—Ç–æ–≤ —Ä–∞–∑–Ω—ã—Ö —Ç–∞—Ä–∏—Ñ–Ω—ã—Ö –ø–ª–∞–Ω–æ–≤ –æ–¥–∏–Ω–∞–∫–æ–≤–∞—è) ") 
```
