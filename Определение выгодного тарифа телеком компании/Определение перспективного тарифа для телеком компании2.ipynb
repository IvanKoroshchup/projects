**Шаг 2.1** <br>
Подготовим данные:
 - Приведем данные к нужным типам; 
 - Найдем и исправим ошибки в данных.

 - Длительность многих звонков 0.0 - оценим, как их отсутствие повлияет на результаты анализа.

```python
#Приведем данные к нужным типам
data_users['reg_date'] = pd.to_datetime(data_users['reg_date'], format='%Y-%m-%d')
data_users['churn_date'] = pd.to_datetime(data_users['churn_date'], format='%Y-%m-%d')
data_calls['call_date'] = pd.to_datetime(data_calls['call_date'], format='%Y-%m-%d')
data_internet['session_date'] = pd.to_datetime(data_internet['session_date'], format='%Y-%m-%d')
data_messages['message_date'] = pd.to_datetime(data_messages['message_date'], format='%Y-%m-%d')

#Округлим звонки и мегабайты до целых значений "вверх"
data_calls['duration'] = data_calls['duration'].apply(np.ceil).astype('int')
data_internet['mb_used'] = data_internet['mb_used'].apply(np.ceil).astype('int')

#Избавимся от дублирующего идексы столбца
data_internet = data_internet[['id','mb_used','session_date','user_id']]

#Удалим "нулевые" звонки, гипотеза: это входящие абоненту, которые никак на его траты не влияют, затем сравним датасеты
data_calls_new = data_calls.query('duration != 0')
```

**Шаг 2.2** <br>
Посчитаем для каждого пользователя:

 - количество сделанных звонков;
 - израсходованных минут разговора по месяцам;
 - количество отправленных сообщений по месяцам;
 - объем израсходованного интернет-трафика по месяцам;
 - помесячную выручку с каждого пользователя (вычтем бесплатный лимит из суммарного количества звонков, сообщений и интернет-трафика; остаток умножем на значение из тарифного плана; прибавим абонентскую плату, соответствующую тарифному плану).

```python
#Посчитаем с разбивкой по месяцам количество сделанных звонков каждым пользователем
data_calls_new['month'] = data_calls_new['call_date'].dt.month
data_users_calls = data_users.merge(data_calls_new, how='outer', on='user_id')
print('===========================================================================================================')
print('количество сделанных звонков')
count_calls = data_users_calls.pivot_table(index = ['user_id','tariff'],columns = 'month',values = 'duration',aggfunc ='count',fill_value = 0)
display(count_calls)

#Посчитаем с разбивкой по месяцам количество израсходованных минут каждым пользователем
print('===========================================================================================================')
print('количество израсходованных минут')
sum_calls = data_users_calls.pivot_table(index = ['user_id','tariff'],columns = 'month',values = 'duration',aggfunc ='sum',fill_value = 0)
display(sum_calls)

#Посчитаем с разбивкой по месяцам количество отправленных СМС
data_messages['month'] = data_messages['message_date'].dt.month
data_users_messages = data_users.merge(data_messages, how='outer', on='user_id')
print('===========================================================================================================')
print('количество отправленных сообщений')
count_messages = data_users_messages.pivot_table(index = ['user_id','tariff'],columns = 'month',values = 'id',aggfunc ='count',fill_value = 0)
display(count_messages)

#Посчитаем объем израсходованного интернет траффика с разбивкой по месяцам
data_internet['month'] = data_internet['session_date'].dt.month
data_users_internet = data_users.merge(data_internet, how='outer', on='user_id')
print('===========================================================================================================')
print('объем израсходованных мегабайтов')
sum_internet = data_users_internet.pivot_table(index = ['user_id','tariff'],columns = 'month',values = 'mb_used',aggfunc ='sum',fill_value = 0)
display(sum_internet)

#Добавим для пользователей колонки месяца подключения (и отключения) тарифа
data_users['reg_month'] = data_users['reg_date'].dt.month
data_users['churn_month'] = data_users['churn_date'].dt.month
data_users_reg = data_users[['user_id','churn_month','reg_month']]
```

  ### Посчитаем доход с каждого клиента по месяцам

```python
#Посчитаем доход с каждого клиента по месяцам
#Соберем одну таблицу объеденив таблицы (сообщения + звонки + мегабайты + пользователи(колонки дата подключения/отключения))
profit_user = sum_calls.join(count_messages, how = 'outer',lsuffix = '_calls', rsuffix='_sms')
sum_internet = sum_internet.add_suffix('_mb')
profit_user = profit_user.join(sum_internet, how = 'outer')
profit_user.reset_index(inplace=True)
profit_user = profit_user.merge(data_users_reg, on = 'user_id', how = 'left')

#Заменим отсутствующие значения в колонке окончания тарифного плана (это нужно для корректности функции)
#В значение valid для функции можно поставить 1000 например, т.к. наши данные все собраны в 2018г, пусть будет равным 12
valid = 12
profit_user = profit_user.fillna({'churn_month':valid})
profit_user = profit_user.fillna(0)

#Для просчета доходности по месяцам напишем функции
def profit_jan (row):
    over_calls = 0
    over_messages = 0
    over_mb = 0
    calls = row['1.0_calls']
    sms = row['1.0_sms']
    mb = row['1.0_mb']
    month_numb = 1
    
    if row['tariff'] == "smart":
        tariff_params = data_tariffs.loc[0]
    elif row['tariff'] == "ultra":
        tariff_params = data_tariffs.loc[1]     
    
    if row['reg_month'] > month_numb or row['churn_month'] < month_numb:
        result = 0
    else:
        if calls > tariff_params['minutes_included']:
            over_calls = (calls - tariff_params['minutes_included'])*tariff_params['rub_per_minute']
        else:
            over_calls = 0
        if sms > tariff_params['messages_included']:
            over_messages = (sms - tariff_params['messages_included'])*tariff_params['rub_per_message']
        else:
            over_messages = 0
        if mb > tariff_params['mb_per_month_included']:
            over_mb = (mb - tariff_params['mb_per_month_included'])*(tariff_params['rub_per_gb']/1024)
        else:
            over_mb = 0
        result = over_calls + over_messages + over_mb + tariff_params['rub_monthly_fee']
        return result
profit_user['profit_jan'] = profit_user.apply(profit_jan, axis = 1)
```

Проверим гипотезы:
 - средняя выручка пользователей тарифов «Ультра» и «Смарт» различается; <br>
Поскольку в нашем случае мы сравниваем средние значения 2х выборок, наша нулевая гипотеза будет звучать так: Среднее двух выборок равно, соответственно выручка от пользователей тарифов смарт и ультра для компании одинаковая.<br>
Альтернативная гипотеза будет звучать так: Среднее двух выборок различается, соответственно выручка от пользователей тарифов смарт и ультра для компании различается<br>

 - средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов<br>
Опять же, мы сравниваем средние значения 2х выборок, наша нулевая гипотеза будет звучать так: Среднее двух выборок равно, соответственно выручка от пользователей городы Москвы и пользователей других городов для компании одинаковая.<br>
Альтернативная гипотеза будет звучать так: Среднее двух выборок различается, соответственно выручка от пользователей городы Москвы и других городов для компании различается

```python
#Разобьем на города Москва/другие города
    #Соединим таблицы годовой выручки и таблицу данных пользователей(нужна колонка с городом)
data_users_city = data_users[['user_id','city']]
profit_user_year_city = profit_user_year.merge(data_users_city, on = 'user_id', how = 'outer')

#Сделаем выборку с годовой выручкой по пользователям из Москвы
profit_user_year_city_moscow = profit_user_year_city\
                .query('city == "Москва"')\
                .drop(['user_id','tariff','city'], axis = 1)\
                .sum(axis = 'columns')

#Сделаем выборку с годовой выручкой по пользователям из других городов
profit_user_year_city_no_moscow = profit_user_year_city\
                .query('city != "Москва"')\
                .drop(['user_id','tariff','city'], axis = 1)\
                .sum(axis = 'columns')

#Сравним среднее выборок выручка Москва/другие города
    #Проверим нормальность распределения выборок Москва/ другие города тестом Шапиро-Уилка
stat_moscov,pvalue_mocsow = st.shapiro(profit_user_year_city_moscow)
stat_no_moscov,pvalue_no_mocsow = st.shapiro(profit_user_year_city_no_moscow)
alpha = .05

if pvalue_mocsow > alpha:
    print('данные в выборке г. Москва распределены нормально (тест Шапиро-Уилка)')
else:
    print('данные в выборке г. Москва распределены не нормально (тест Шапиро-Уилка)')

if pvalue_no_mocsow > alpha:
    print('данные в выборке по городам кроме Москвы распределены нормально (тест Шапиро-Уилка)')
else:
    print('данные в выборке по городам кроме Москвы распределены не нормально (тест Шапиро-Уилка)')    
print()

#Видим что обе выборки распределены не нормально, используем тест Левена для оценки равенства дисперсий
alpha = .05
stat, pval = st.levene(profit_user_year_city_moscow, profit_user_year_city_no_moscow)
if pval > alpha:
    print('дисперсия выборок Москва / другие города - равна (тест Левена)')
else:
    print('дисперсия выборок Москва / другие города - не равна (тест Левена)')    
print()

#Зная, что дисперсии равны, вписываем параметр equal_var
alpha = .05
results = st.ttest_ind(profit_user_year_city_moscow, profit_user_year_city_no_moscow, equal_var = True)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу (выручка от абонентов Москвы и других городах одинаковая)")
else:
    print("Не получилось отвергнуть нулевую гипотезу (выручка от абонентов Москвы и других городов одинаковая)") 
print('============================================================================================')


#Сравним среднее выборок выручка тарифов смарт/ультра
    #Узнаем годовую выручку по каждому пользователю тарифов(выборка по тарифу "смарт" и ультра у нас уже есть)
profit_user_year_smart_sum = profit_user_year_smart.sum(axis = 'columns')
profit_user_year_ultra_sum = profit_user_year_ultra.sum(axis = 'columns') 
    
    #Проверим нормальность распределения выборок Москва/ другие города тестом Шапиро-Уилка
stat_smart,pvalue_smart = st.shapiro(profit_user_year_smart_sum)
stat_ultra,pvalue_ultra = st.shapiro(profit_user_year_ultra_sum)
alpha = .05
    
if pvalue_smart > alpha:
    print('данные в выборке тариф "смарт" распределены нормально (тест Шапиро-Уилка)')
else:
    print('данные в выборке тариф "смарт" распределены не нормально (тест Шапиро-Уилка)')

if pvalue_ultra > alpha:
    print('данные в выборке тариф "ультра" распределены нормально (тест Шапиро-Уилка)')
else:
    print('данные в выборке тариф "ультра" распределены не нормально (тест Шапиро-Уилка)')    
print()

#Видим что обе выборки распределены не нормально, используем тест Левена для оценки равенства дисперсий
alpha = .05
stat_tariffs, pval_tariffs = st.levene(profit_user_year_smart_sum, profit_user_year_ultra_sum)
if pval_tariffs > alpha:
    print('дисперсия выборок Москва / другие города - равна (тест Левена)')
else:
    print('дисперсия выборок Москва / другие города - не равна (тест Левена)')    
print()    
    
#Зная, что дисперсии равны, вписываем параметр equal_var
alpha = .05
results = st.ttest_ind(profit_user_year_smart_sum, profit_user_year_ultra_sum, equal_var = False)
if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу: (выручка от абонентов разных тарифных планов одинаковая)")
else:
    print("Не получилось отвергнуть нулевую гипотезу: (выручка от абонентов разных тарифных планов одинаковая) ") 
```
