### Шаг 2. Предобработка данных

Много пропущеных значений, как незаполненных самими пользователями, так и незаполненых сервисом.<br>
1)Заполняем столбец балконы на ноль, там где количество не указано, вероятно, что люди сами не заполнили эти строки так как нет балконов.<br>
2)Заполняем столбец с аппартаментами (там, где не указано, предполагаем что не аппартаменты"<br>
3)Удалим строки с отсутствующим значением населенного пункта, таких 49 штук<br>
4)Изменяем типы данных в столбцах:<br>
 - first_day_exposition - на дату
 - is_apartment - в булев тип (значение либо да/либо нет)



```python
#Заполняем столбец балконы на ноль, там где количество не указано
data['balcony'] = data['balcony'].fillna(0)

#Заполняем столбец на квартиры, там где отдельно не указано, что это аппартаменты
data['is_apartment'] = data['is_apartment'].fillna(False)

#Удаляем строчки без указания этажности здания
data.dropna(subset = ['floors_total'], inplace = True)

#Удаляем строчки без названия населенного пункта
data.dropna(subset = ['locality_name'], inplace = True)

#Изменяем типы данных в столбцах
data['is_apartment'] = data['is_apartment'].astype('bool')
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format='%Y.%m.%dT%H:%M:%S')

#Переведем стоимость в миллионы, так удобнее
data['last_price'] = data['last_price'] / 1000000

#Заполним пропуски в столбце с параметрами кухонь
kitchen_koff = (data['kitchen_area']/data['total_area']).mean()
kitchen_koff_na = data.loc[:,'total_area']*kitchen_koff
data['kitchen_area'] = data['kitchen_area'].fillna(kitchen_koff_na)

#Аналогично поступим с пустыми значениями в жилой площади
living_koff = (data['living_area']/data['total_area']).mean()
living_koff_na = data.loc[:,'total_area']*living_koff
data['living_area'] = data['living_area'].fillna(living_koff_na)

#Заполним пропуски в столбце с высотой потолков
for ceiling_na in data.floors_total.unique():
    ceiling = data.query('floors_total == @ceiling_na').ceiling_height.median()
    data.loc[data.floors_total == ceiling_na, 'ceiling_height'] = data.loc[data.floors_total == ceiling_na, 'ceiling_height'].fillna(ceiling)  
```

### Шаг 3. Посчитаем и добавим в таблицу
 - Цену за 1 кв.м.
 - День недели, месяц и год публикации объявления
 - Этаж квартиры; варианты — первый, последний, другой;
 - Соотношение жилой и общей площади, 
 - Соотношение площади кухни к общей площади.

```python
#Цена за кв. метр.
data['price_per_meter'] = data['last_price']/data['total_area']

#Добавим столбец - день недели публикации объявления (у нас уже есть столбец, где указан месяц и год публикации объявления)
data['publication_weekday'] = data['first_day_exposition'].dt.weekday

#Добавим столбец - Этаж первый, последний, другой
def floor_type (row):
    floors_total = row['floors_total']
    floor = row['floor']
    if floor == 1:
        return 'первый'
    if floor == floors_total:
        return 'последний'
    return 'другой'
row_values = [1, 1]
row_columns = ['floor','floors_total']
row = pd.Series(data=row_values, index = row_columns)   
data['floor_status'] = data.apply(floor_type, axis=1)

#Добавим столбец - Соотношение жилой и общей площади
data['living_area/total_area'] = data['living_area']/data['total_area']

#Добавим столбец - Соотношение площади кухни к общей площади
data['kitchen_area/total_area'] = data['kitchen_area']/data['total_area']
```

**Какие факторы больше всего влияют на стоимость квартиры?** Изучим, зависит ли цена от площади, числа комнат, </br>
   удалённости от центра. Изучим зависимость цены от того, на каком этаже расположена квартира: первом, последнем 
   или другом. Также изучим зависимость от даты размещения: дня недели, месяца и года.</br>

```python
#Добавим столбец с расстоянием, округленным до км
data_good['cityCenters_nearest_km'] = data_good['cityCenters_nearest']/ 1000
data_good.dropna(subset = ['cityCenters_nearest_km'], inplace = True)
data_good['cityCenters_nearest_km'] = data_good['cityCenters_nearest_km'].astype('int64')

#Для корректности сравнения возбмем квартиры в одном регионе
#display(data_good)
data_good_piter = data_good.query('locality_name == "Санкт-Петербург"')
data_good_piter = data_good[['last_price', 'total_area', 'first_day_exposition','rooms','floor_status', 'cityCenters_nearest_km','publication_weekday','ceiling_height']]

#Общая корреляция между собой
sns.heatmap(data_good_piter.corr(), annot = True)

#Влияние на цену: площадь
data_good_piter.plot(x='total_area',y='last_price',kind='scatter',alpha = 0.2,xlim=(0,100),ylim=(0,12),title="Влияние на цену: площадь")
plt.show()

#Влияние на цену: количество комнат
data_good_piter.plot(kind='scatter',alpha = 0.02,xlim=(0,8),ylim=(0,20),x='rooms',y='last_price',title="Влияние на цену:количество комнат")
plt.show()

print('Вывод: на стоимость квартир в большей степени влияет общая площадь \
(кофф. 0,62),количество комнат (кофф. 0,33) и расстояние до центра (кофф. 0,17), зависимости от дня публикации -нет')

#Влияние на цену: этаж первый, последний, другой
ax = data_good_piter.query('floor_status == "первый"').sort_values(by='last_price')\
    .plot(xlim=(0,20), y = 'last_price',histtype = 'step',kind ='hist',bins= 800,label = 'первый этаж',figsize = (15,4))
data_good_piter.query('floor_status == "последний"')\
    .plot(y = 'last_price',histtype = 'step',kind ='hist',bins= 800, ax=ax,label = 'последний этаж')
data_good_piter.query('floor_status == "другой"')\
    .plot(y = 'last_price',kind ='hist',histtype = 'step',bins= 800, ax=ax,label = 'другой',legend = True)

print('Вывод: самые дешевые квартиры на первом этаже, на втором месте -последний, самые дорогие на остальных')
```
